Figure 1: Number of Passengers by Family Size (Grouped by Survival)

The first chart shows how Family Size relates to survival. The largest group by far were passengers travelling alone (Family Size = 1), with over 350 people. This group suffered the highest mortality, as the number of non-survivors is roughly double that of survivors. Travelling alone likely meant fewer social ties and less support during evacuation, which may have reduced the urgency or ability to reach safety.

Interestingly, survival rates rise among those with small family sizes, particularly groups of two or three, where children were often present. Families with young children appear to have been prioritised in evacuation, reflecting the “women and children first” policy. These groups had a more balanced survival distribution, suggesting that being accompanied by dependents improved one’s chances.

However, for larger families (four or more), survival drops sharply again. This may be because larger families found it difficult to stay together or move quickly, and the sheer number of dependents made evacuation more complicated. Overall, Family Size is a valuable predictor, capturing both the disadvantage of being alone and the influence of dependents on survival outcomes.

Figure 2: Number of Passengers / Survival by Sex and Passenger Class

The second chart examines survival by Sex and Passenger Class, and the differences are striking. Female passengers in first and second class (F_C1, F_C2) had very high survival rates, while men in third class (M_C3) accounted for the majority of deaths. This aligns closely with the social norms and evacuation priorities of the time, where women and children in higher classes were given preference for lifeboats.

Among women, survival declines steadily from first to third class, and a similar pattern appears among men. Even within each sex, class remains decisive: wealthier passengers were generally located closer to the lifeboats and had quicker access to the decks. Third class passengers, particularly men, were at a significant disadvantage due to location, language barriers, and overcrowding.

The interaction between Sex and Passenger Class is therefore crucial. Modelling these features together, for example, by creating a combined variable such as “Sex_Class”, can better capture the social and structural inequalities that determined survival.
